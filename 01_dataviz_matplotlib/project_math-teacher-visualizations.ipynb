```python
from matplotlib import pyplot as plt
%matplotlib inline 
```

## 1. Bar Chart with Error

```python
#Data
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]

#Create a figure of width 10 and height 8
plt.figure(figsize=(10,8))

#Plot blue bars, add error bars of cap size 3 and heights corresponding to error
plt.bar(range(len(past_years_averages)),past_years_averages,capsize=3,yerr=error)

#Set the axis to go from -0.5 to 6.5 on the x-axis and 70 to 95 on the y-axis
plt.axis([-0.5, 6.5, 70, 95])

#Create an ax object, and set x-axis labels to be the years
ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

#Add labels and title
plt.title('Final Exam Averages')
plt.ylabel('Test average')
plt.xlabel('Year')

#Show plot
plt.show()
```

## 2. Side By Side Bars


```python
#Data
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

school_a_x = [0.8, 2.8, 4.8, 6.8, 8.8]
school_b_x = [1.6, 3.6, 5.6, 7.6, 9.6]

#Create figure width of 10 and 8
plt.figure(figsize=(10,8))

#Create axes
ax = plt.subplot()

#Plot middle school a and b on same plot
plt.bar(school_a_x,middle_school_a)
plt.bar(school_b_x,middle_school_b)

#Create a new list of x-values called middle_x between a_x and b_x, in order to place the x-ticks
middle_x = [(a+b)/2 for (a,b) in zip(school_a_x,school_b_x)]

#Set x-ticks to be the middle_x list
ax.set_xticks(middle_x)

#Set x-tick labels to be unit topics
ax.set_xticklabels(unit_topics)

#Create legend
plt.legend(["Middle School A", "Middle School B"])

#Create title 
plt.title("Test Averages On Different Units")
plt.xlabel("Unit")
plt.ylabel("Test Average")

plt.show()
```

## 3. Stacked Bars

```python
#Data
import numpy as np

#Data
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)

#Denoting the heights each grade should start
c_bottom = np.add(As, Bs)
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

#Create a figure of width 10 and height 8
plt.figure(figsize=(10,8))

#Create the bar plot
plt.bar(x,As)
plt.bar(x,Bs,bottom=As)
plt.bar(x,Cs,bottom=c_bottom)
plt.bar(x,Ds,bottom=d_bottom)
plt.bar(x,Fs,bottom=f_bottom)

#Create set of axes
ax = plt.subplot()

#set x-ticks
ax.set_xticks(x)

#Set x-tick labels
ax.set_xticklabels(unit_topics)

#Set Legend
plt.legend(["As","Bs","Cs","Ds","Fs"])

#Set title and axes labels
plt.title("Grade distribution")
plt.xlabel("Unit")
plt.ylabel("Number of students")
plt.show()
```

## 6. Line with Shaded Error


```python
#data
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

#Create initial plot
plt.plot(exam_scores,hours_reported,linewidth=2)

#Create errors assuming 20%
hours_lower_bound = [i*0.8 for i in hours_reported]
hours_upper_bound = [i*1.2 for i in hours_reported]

#Fill between for errors shading
plt.fill_between(exam_scores,hours_lower_bound,hours_upper_bound,alpha=0.2)

#Title and labels
plt.title("Time spent studying vs. final exam scores")
plt.xlabel('Score')
plt.ylabel('Hours studying (self-reported)')

plt.show()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

```python
#Create a figure width 12, height 8, and 2 subplots 1 row
plt.figure(figsize=(12,8))

#Left subplot: page visits over time
ax1 = plt.subplot(1,2,1)

#create a list of x values number 1-12 to represent months
x_values = range(len(months))

#create the line plot with round marker for distinction
plt.plot(x_values,visits_per_month,marker = 'o')

#Label the x and y-axis
plt.xlabel('Months')
plt.ylabel('Visits Over Time')

#Set x-ticks to be x-values
ax1.set_xticks(x_values)

#Change xtick labels to be month names
ax1.set_xticklabels(months)

#Add title to left subplot
plt.title('Page Visits By Month')

#Right subplot: plotting multiple lime species
ax2 = plt.subplot(1,2,2)

#Create plots of the different lime species against the x-values
plt.plot(x_values,key_limes_per_month,color = 'green', marker = 'o')
plt.plot(x_values, persian_limes_per_month, color = 'orange',marker = 'o')
plt.plot(x_values, blood_limes_per_month, color = 'red', marker = 'o')
plt.legend(['key limes','persian limes','blood limes'], loc = 1)
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.title('Lime Species Sales by Month')

plt.savefig('lime_trends.png')
```

```python
#Plot page visits and Total Lime Sales on same plot

#Create total limes list using list comprehensions and zip
total_lime_sales = [sum(x) for x in zip(key_limes_per_month, blood_limes_per_month,persian_limes_per_month)]

#Create a figure width 12, height 8, and 2 subplots 1 row
fig, ax1 = plt.subplots()

#create a list of x values number 1-12 to represent months
x_values = range(len(months))

#create the line plots with round marker for distinction, with page visits and total lime sales on distinct lines
ax2 = ax1.twinx()

ax1.plot(x_values,visits_per_month,marker = 's',color = 'blue', linestyle = '--')
ax2.plot(x_values, total_lime_sales, marker = 'o', color = 'orange')

#Label the x and y-axes appropriately
ax1.set_xlabel('Months')
ax1.set_ylabel('Page Visits')
ax2.set_ylabel('Total Lime Sales')

#Set x-ticks to be x-values
ax1.set_xticks(x_values)

#Change xtick labels to be month names
ax1.set_xticklabels(months)

#add legends
ax1.legend(['page visits'], loc = 0)
ax2.legend(['total lime sales'], loc = 0)

#Add title to left subplot
plt.title('Page Visits vs. Lime Sales By Month') 

plt.savefig('lime_and_sales_trends.png')
```
